```python
import matplotlib.pyplot as plt
import numpy as np

models = ["DistilBERT", "LLaMA-7B", "GPT-3", "GPT-4", "Falcon-40B", "BERT-Base", "BLOOM", "XLM-RoBERTa-Base", "DeepSeek-R1-0528", "Mixtral-8x22B"]
co2_emissions = [200, 15000, 500000, 800000, 375000, 635, 25000, 635, 500000, 200000]  # kg CO2
parameters = [66, 7000, 175000, 1000000, 40000, 110, 176000, 279, 685000, 141000]  # millions

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.bar(models, co2_emissions, color=['green', 'orange', 'red', 'darkred'])
ax1.set_title('CO2 Emissions During Training')
ax1.set_ylabel('CO2 (kg)')
ax1.tick_params(axis='x', rotation=45)

# Parameters vs emissions
ax2.scatter(parameters, co2_emissions, s=100)
for i, model in enumerate(models):
    ax2.annotate(model, (parameters[i], co2_emissions[i]))
ax2.set_xlabel('Parameters (millions)')
ax2.set_ylabel('CO2 Emissions (kg)')
ax2.set_title('Model Size vs Environmental Impact')

plt.tight_layout()
plt.show()
```
